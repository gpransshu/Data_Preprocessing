```python
import pandas as pd
import numpy as np
```

```python
d1 = {'c1':[1,2,3],
      'c2':[10,11,12]}
d2 = {'c3':[4,5,6],
      'c4':[13,14,15]}
d3 = {'c5':[7,8],
      'c6':[16,17]}
```

```python
df1 = pd.DataFrame(d1)
df1
```

```python
df2 = pd.DataFrame(d2)
df2
```

```python
df3 = pd.DataFrame(d3)
df3
```

```python
#concatenation
pd.concat([df1,df2])
```

```python
#merging
d1 = {'c1':[1,2,3],
      'c2':[10,5,12]}
d2 = {'c2':[4,5,6],
      'c4':[13,14,15]}
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
df2
```

```python
pd.concat([df1,df2])
```

```python
#common values
pd.merge(df1,df2,how = 'inner', on = 'c2')
```

```python
#Create two dfs such that with the help of inner join we're gonna return 3 rows
a = {'p':[10,20,30,40],
      'q':[40,50,60,70]}
b = {'q':[50,60,70,80],
      'r':[80,90,100,110]}
a = pd.DataFrame(a)
b= pd.DataFrame(b)
```

```python
a
```

```python
b
```

```python
pd.merge(a,b,how = 'inner', on = 'q')
```

```python
pd.merge(a,b,how = 'left', on = 'q')
```
